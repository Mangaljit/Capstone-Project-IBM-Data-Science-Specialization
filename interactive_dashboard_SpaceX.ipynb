```python
# Run this app with `python app.py` and
# visit http://127.0.0.1:8050/ in your web browser.

import pandas as pd
import os
from dash import Dash
import dash_html_components as html
import dash_core_components as dcc
from dash.dependencies import Input, Output
import plotly.express as px

pwd = os.getcwd()
data = pd.read_csv(pwd + "/spacex_launch_dash.csv")
maxload = data['Payload Mass (kg)'].max()
minload = data['Payload Mass (kg)'].min()

app = Dash(__name__)
server = app.server

all_sites = "All Sites"
sites = data['Launch Site'].unique().tolist()
sites.append(all_sites)

app.layout = html.Div(children=[html.H1("SpaceX Dashboard Showing Different Launch-Site's Success rate",
                                        style={'textAlign': 'center', 'color': '#503D36',
                                               'font-size': 40}),
                                dcc.Dropdown(id='dropdown',options=sites,placeholder='Select a Launch Site here', searchable = True , value = 'All Sites'),
                                html.Br(),
                                html.Div(dcc.Graph(id='success-pie-chart')),
                                html.Br(),
                                html.P("Slide and Select Payload Range in Kg:"),
                                dcc.RangeSlider(
                                    id='slider',
                                    min=0,
                                    max=10000,
                                    step=1000,
                                    marks = {
                                            0: '0 kg',
                                            1000: '1000 kg',
                                            2000: '2000 kg',
                                            3000: '3000 kg',
                                            4000: '4000 kg',
                                            5000: '5000 kg',
                                            6000: '6000 kg',
                                            7000: '7000 kg',
                                            8000: '8000 kg',
                                            9000: '9000 kg',
                                            10000: '10000 kg'
                                    },
                                    value=[minload,maxload]
                                ),
                                html.Div(dcc.Graph(id='success-payload-scatter-chart')),
                                ])

@app.callback(
     Output(component_id='success-pie-chart',component_property='figure'),
     [Input(component_id='dropdown',component_property='value')]
)
def update_graph(dropdown):
    if (dropdown == 'All Sites'):
        df  = data[data['class'] == 1]
        fig = px.pie(df, names = 'Launch Site', title = 'Success Launch Rate For All Sites')
    else:
        df  = data.loc[data['Launch Site'] == dropdown]
        fig = px.pie(df, names = 'class',title = 'Success Launch Rate for Site '+dropdown)
    return fig

@app.callback(
     Output(component_id='success-payload-scatter-chart',component_property='figure'),
     [Input(component_id='dropdown',component_property='value'),Input(component_id="slider", component_property="value")]
)
def update_scattergraph(dropdown,slider):
    if dropdown == 'All Sites':
        low, high = slider
        df  = data
        mask = (df['Payload Mass (kg)'] > low) & (df['Payload Mass (kg)'] < high)
        fig = px.scatter(
            df[mask], x="Payload Mass (kg)", y="class",
            color="Booster Version",
            #size='Payload Mass (kg)',
            hover_data=['Payload Mass (kg)'], title = 'Correlation Between Payload and Success For All Sites')
    else:
        low, high = slider
        df  = data.loc[data['Launch Site'] == dropdown]
        mask = (df['Payload Mass (kg)'] > low) & (df['Payload Mass (kg)'] < high)
        fig = px.scatter(
            df[mask], x="Payload Mass (kg)", y="class",
            color="Booster Version",
            #size='Payload Mass (kg)',
            hover_data=['Payload Mass (kg)'],title = 'Correlation Between Payload and Success For All Sites')
    return fig

if __name__ == '__main__':
    app.run_server(debug=False)


```
